```python
import numpy as np
import matplotlib.pyplot as plt

# Load the 2D FES data (assuming 1st, 2nd, and 3rd columns are phi, psi, and free energy)
### 7 AM, Oct 17, 2024 ###
fes_2d_data = np.loadtxt('fes_2d.dat')

# Reshape the data into 2D grid
phi_values = np.unique(fes_2d_data[:, 0])
psi_values = np.unique(fes_2d_data[:, 1])
free_energy = fes_2d_data[:, 2].reshape(len(phi_values), len(psi_values))

# Create the plot with imshow and contours
plt.figure(figsize=(8, 6))
plt.imshow(free_energy, extent=[-np.pi, np.pi, -np.pi, np.pi], origin='lower', cmap='viridis', aspect='auto')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.contour(phi_values, psi_values, free_energy, colors='white', levels=10)
plt.xlabel('Phi (radians)')
plt.ylabel('Psi (radians)')
plt.title('2D Free Energy Surface (Phi vs Psi)')
plt.savefig('Step3_Figures/2d_fes_contours.png')
plt.show()

### Wow... this looks so coolllll, Thank you ChatGPT and the tutorial --- Jialong ###
```
