```python
import numpy as np
import matplotlib.pyplot as plt

# Load COLVAR file
dataA = np.loadtxt('stateA_simulation/COLVAR')
dataB = np.loadtxt('stateB_simulation/COLVAR')

# Extract phi and psi columns (assumed 2nd and 3rd columns)
timeA = dataA[:, 0]
phiA = dataA[:, 1]
psiA = dataA[:, 2]

timeB = dataB[:, 0]
phiB = dataB[:, 1]
psiB = dataB[:, 2]

# Plot phi vs time for state A and B
plt.plot(timeA, phiA, label="State A Phi")
plt.plot(timeB, phiB, label="State B Phi")
plt.xlabel('Time (ps)')
plt.ylabel('Phi (degrees)')
plt.legend()
plt.title('Phi vs Time')
plt.savefig('Step4_Figures/phi_vs_time.png')  # Save figure
plt.show()

# Plot psi vs time for state A and B
plt.plot(timeA, psiA, label="State A Psi")
plt.plot(timeB, psiB, label="State B Psi")
plt.xlabel('Time (ps)')
plt.ylabel('Psi (degrees)')
plt.legend()
plt.title('Psi vs Time')
plt.savefig('Step4_Figures/psi_vs_time.png')  # Save figure
plt.show()
```

```python
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt

# Load COLVAR data
colvar_dataA = np.loadtxt('stateA_simulation/COLVAR')

# Extract phi and psi from COLVAR (assuming they are in the second and third columns)
phi_colvar = colvar_dataA[:, 1]  # Phi from COLVAR
psi_colvar = colvar_dataA[:, 2]  # Psi from COLVAR

# Load the trajectory
trajA = md.load('stateA_simulation/traj_comp.xtc', top='../Inputs/dialaA.pdb')

# Compute phi and psi dihedrals from MDTraj
phi_A_mdtraj = md.compute_phi(trajA)[1]
psi_A_mdtraj = md.compute_psi(trajA)[1]

# Plot the data from both COLVAR and MDTraj on the same plot for comparison
plt.figure(figsize=(8,6))

# Scatter plot of COLVAR data with dots ('.') in blue
plt.scatter(phi_colvar, psi_colvar, label="COLVAR Phi vs Psi", alpha=0.5, color='blue', marker='x')

# Scatter plot of MDTraj data with circles ('o') in red
plt.scatter(phi_A_mdtraj, psi_A_mdtraj, label="MDTraj Phi vs Psi", alpha=0.5, color='red', marker='o')

# Add labels and legend
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.legend()
plt.title('Comparison of Dihedrals for State A: COLVAR vs MDTraj (X and Circles)')

# Save the figure to the Step4_Figures directory
plt.savefig('Step4_Figures/comparison_phi_psi_state_A.png')

plt.show()
```
