```python
import numpy as np
import matplotlib.pyplot as plt

### Visualizing the COLVAR File ###

# Load the COLVAR data from the renamed file
colvar_file = 'ala2_metad_biasfactor10_pace500_height1.2_sigma0.3_10ns.colvar.txt'
colvar_data = np.loadtxt(colvar_file)

# Extract time, phi, and psi (assuming the 1st, 2nd, and 3rd columns are time, phi, psi)
time = colvar_data[:, 0]  # Time is in the first column
phi = colvar_data[:, 1]   # Phi is in the second column
psi = colvar_data[:, 2]   # Psi is in the third column

# Plot every 100th point to reduce the density of the graph
step = 100  # Adjust this number to reduce the number of points plotted

# Plot Phi (ϕ) vs Time
plt.figure(figsize=(8, 6))
plt.plot(time[::step], phi[::step], label='Phi (ϕ)', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Phi (ϕ) (degrees)')
plt.title('Phi (ϕ) vs Time')
plt.legend()
plt.savefig('Step3_Figures/phi_vs_time_reduced.png')  # Save figure
plt.show()

# Plot Psi (ψ) vs Time
plt.figure(figsize=(8, 6))
plt.plot(time[::step], psi[::step], label='Psi (ψ)', color='red')
plt.xlabel('Time (ps)')
plt.ylabel('Psi (ψ) (degrees)')
plt.title('Psi (ψ) vs Time')
plt.legend()
plt.savefig('Step3_Figures/psi_vs_time_reduced.png')  # Save figure
plt.show()

### Visualizing the HILLS File ###

# Load the HILLS data from the renamed file
hills_file = 'ala2_metad_biasfactor10_pace500_height1.2_sigma0.3_10ns.hills.txt'
hills_data = np.loadtxt(hills_file)

# Extract time and Gaussian height (assuming 1st and 4th columns are time and height)
time_hills = hills_data[:, 0]  # Time is in the first column
gaussian_height = hills_data[:, 3]  # Height is in the fourth column

# Plot Gaussian Height vs Time
plt.figure(figsize=(8, 6))
plt.plot(time_hills, gaussian_height, label='Gaussian Height', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Gaussian Height (kJ/mol)')
plt.title('Gaussian Height vs Time')
plt.legend()
plt.savefig('Step3_Figures/gaussian_height_vs_time.png')  # Save figure
plt.show()
```
