```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from end_to_end.xvg
time, distance = np.loadtxt('end_to_end.xvg', comments=['@', '#'], unpack=True)

# Load the data from gyrate.xvg (only the first two columns: time and Rg)
time_g, rg = np.loadtxt('gyrate.xvg', comments=['@', '#'], usecols=(0, 1), unpack=True)

# Plot end-to-end distance
plt.figure()
plt.plot(time, distance, label='End-to-end Distance')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (nm)')
plt.title('End-to-end Distance vs Time')
plt.legend()
plt.savefig('../1c_1V1D_AMBER_FIGURES/end_to_end_distance.png')

# Plot radius of gyration
plt.figure()
plt.plot(time_g, rg, label='Radius of Gyration', color='orange')
plt.xlabel('Time (ps)')
plt.ylabel('Radius of Gyration (nm)')
plt.title('Radius of Gyration vs Time')
plt.legend()
plt.savefig('../1c_1V1D_AMBER_FIGURES/radius_of_gyration.png')
```
