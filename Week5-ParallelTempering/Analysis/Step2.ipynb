```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the trajectory and topology files
traj = md.load('../Data/md_15ns.xtc', top='../Data/md_15ns.gro')  # Use .gro as the topology instead of .tpr

# Compute the phi and psi dihedral angles
phi_indices, phi_angles = md.compute_phi(traj)
psi_indices, psi_angles = md.compute_psi(traj)

# Convert radians to degrees for easier interpretation
phi_angles_deg = np.rad2deg(phi_angles)
psi_angles_deg = np.rad2deg(psi_angles)

# Plot phi angles vs time
plt.figure(figsize=(10, 5))
plt.plot(traj.time, phi_angles_deg, label='Phi (degrees)', color='blue')
plt.xlabel('Time (ps)')
plt.ylabel('Phi (degrees)')
plt.title('Phi Dihedral Angles vs Time')
plt.legend()
plt.grid(True)
plt.savefig('../Figures/phi_vs_time.png')  # Save the figure
plt.show()

# Plot psi angles vs time
plt.figure(figsize=(10, 5))
plt.plot(traj.time, psi_angles_deg, label='Psi (degrees)', color='green')
plt.xlabel('Time (ps)')
plt.ylabel('Psi (degrees)')
plt.title('Psi Dihedral Angles vs Time')
plt.legend()
plt.grid(True)
plt.savefig('../Figures/psi_vs_time.png')  # Save the figure
plt.show()
```
