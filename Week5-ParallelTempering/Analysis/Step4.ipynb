```python
import numpy as np
import matplotlib.pyplot as plt

# Load potential energy data for each temperature
energy_300 = np.loadtxt('../Inputs/T300/potential_energy.xvg', comments=['#', '@'])
energy_400 = np.loadtxt('../Inputs/T400/potential_energy.xvg', comments=['#', '@'])
energy_600 = np.loadtxt('../Inputs/T600/potential_energy.xvg', comments=['#', '@'])

# Extract time and energy columns (assuming time is in column 0 and energy in column 1)
energy_300_vals = energy_300[:, 1]
energy_400_vals = energy_400[:, 1]
energy_600_vals = energy_600[:, 1]

# Plot the histograms
plt.figure(figsize=(8, 6))
plt.hist(energy_300_vals, bins=50, alpha=0.5, label='T=300K')
plt.hist(energy_400_vals, bins=50, alpha=0.5, label='T=400K')
plt.hist(energy_600_vals, bins=50, alpha=0.5, label='T=600K')
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Frequency')
plt.title('Potential Energy Histograms for T=300K, T=400K, and T=600K')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save and show the figure
plt.savefig('../Figures/potential_energy_histograms.png')
plt.show()
```
