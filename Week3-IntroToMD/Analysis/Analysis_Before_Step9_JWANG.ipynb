```python
import matplotlib.pyplot as plt
import numpy as np

file_path = 'pressure.xvg'

data = np.loadtxt(file_path, comments=["@", "#"])

# Extract time (first column) and pressure (second column)
time = data[:, 0]
pressure = data[:, 1]

# Calculate a simple moving average with a window of 10 points
window_size = 10
running_avg = np.convolve(pressure, np.ones(window_size)/window_size, mode='valid')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, pressure, label='Pressure (bar)', color='black')
plt.plot(time[window_size-1:], running_avg, label='10-ps Running Avg', color='red')

plt.xlabel('Time (ps)')
plt.ylabel('Pressure (bar)')
plt.title('Pressure during NPT Equilibration')
plt.legend()

figure_path = '../Figures/Pressure_during_NPT_Equilibration.png'
plt.savefig(figure_path)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

file_path = 'density.xvg'
data = np.loadtxt(file_path, comments=["@", "#"])

time = data[:, 0]
density = data[:, 1]

window_size = 10
running_avg = np.convolve(density, np.ones(window_size)/window_size, mode='valid')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, density, label='Density (kg/m^3)', color='black')
plt.plot(time[window_size-1:], running_avg, label='10-ps Running Avg', color='red')

plt.xlabel('Time (ps)')
plt.ylabel('Density (kg m$^{-3}$)')
plt.title('Density during NPT Equilibration')
plt.legend()

figure_path = '../Figures/Density_during_NPT_Equilibration.png'
plt.savefig(figure_path)

plt.show()
```
