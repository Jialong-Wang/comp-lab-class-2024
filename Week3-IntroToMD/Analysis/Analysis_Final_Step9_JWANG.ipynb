```python
import matplotlib.pyplot as plt
import numpy as np

time_equil, rmsd_equil = np.loadtxt('../Analysis/rmsd.xvg', comments=['#', '@'], unpack=True)
time_xtal, rmsd_xtal = np.loadtxt('../Analysis/rmsd_xtal.xvg', comments=['#', '@'], unpack=True)

# Plotting
plt.figure(figsize=(6, 5))
plt.plot(time_equil, rmsd_equil, label="Ref: Equilibrated", color='black')
plt.plot(time_xtal, rmsd_xtal, label="Ref: Crystal", color='red')

# Labels and Title
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD\n1AKI, Backbone')

# Limits and grid
plt.ylim(0, 0.4)  # Adjusting y-axis limit for better window control
plt.xlim(0, 10.2)    # Setting x-axis limits

# Adding ticks that match the example plot
plt.xticks(np.arange(0, 10.1, step=1))
plt.yticks(np.arange(0, 0.5, step=0.05))

# Adding a tight layout to avoid cutting off elements
plt.tight_layout()

# Legend
plt.legend()

# Show plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Function to read the xvg file and extract data
def read_xvg(file_path):
    times = []
    gyrate_values = []
    with open(file_path, 'r') as file:
        for line in file:
            # Ignore comments and headers in the file
            if line.startswith(('#', '@')):
                continue
            # Split each line into time and gyrate value
            parts = line.split()
            times.append(float(parts[0]))
            gyrate_values.append(float(parts[1]))
    return times, gyrate_values

# Function to plot data
def plot_gyrate(times, gyrate_values):
    plt.figure(figsize=(10, 6))
    plt.plot(times, gyrate_values, label="Radius of Gyration")
    plt.xlabel('Time (ps)')
    plt.ylabel('Radius of Gyration (nm)')
    plt.title('Radius of Gyration vs Time')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    # Save the plot
    plt.savefig('gyration_plot.png')
    plt.show()

# Main function to read file and plot
if __name__ == "__main__":
    file_path = "../Analysis/gyrate.xvg"  # Adjust this path as necessary
    times, gyrate_values = read_xvg(file_path)
    plot_gyrate(times, gyrate_values)
```
